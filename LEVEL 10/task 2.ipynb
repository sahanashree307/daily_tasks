```python
import pandas as pd
from scipy import stats
import numpy as np
```

```python
train_df = pd.read_csv(r"C:\Users\sahan\Downloads\titanic_train.csv")
train_df
```

```python
train_df.head()
```

```python
survived_age = train_df[train_df['Survived'] == 1]['Age'].dropna()
not_survived_age = train_df[train_df['Survived'] == 0]['Age'].dropna()
t_stat, p_value = stats.ttest_ind(survived_age, not_survived_age)
print("T-Test: Comparing Age of Survivors vs Non-Survivors")
print(f"T-statistic = {t_stat:.3f}, p-value = {p_value:.3f}")
```

```python
from statsmodels.stats.proportion import proportions_ztest
import numpy as np
```

```python
female_surv = train_df[(train_df['Sex'] == 'female') & (train_df['Survived'] == 1)].shape[0]
male_surv = train_df[(train_df['Sex'] == 'male') & (train_df['Survived'] == 1)].shape[0]
```

```python
female_total = train_df[train_df['Sex'] == 'female'].shape[0]
male_total = train_df[train_df['Sex'] == 'male'].shape[0]
```

```python
counts = [female_surv, male_surv]
totals = [female_total, male_total]

z, p = proportions_ztest(counts, totals)
print(f"Z-Test: z = {z:.2f}, p = {p:.3f}")
```

```python
survival_by_gender = train_df.groupby('Sex')['Survived'].value_counts(normalize=True).unstack()
print(survival_by_gender)
```
