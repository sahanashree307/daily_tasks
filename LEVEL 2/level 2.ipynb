```python
#task 3
def multiply(a, b):
    return a * b
```

```python
arr = [1, 2, 3, 4, 5, 6]
num = 2
mapped_result = list(map(lambda x: multiply(x, num), arr))
```

```python
filtered_result = list(filter(lambda x: x % 2 == 0, mapped_result))
```

```python
print("Mapped Result:", mapped_result)
print("Filtered Result:", filtered_result) 
```
